```python
import meshio
import numpy as np

m = meshio.read("chain.msh")

print([c.type for c in m.cells])

tetra_cells = None
for c in m.cells:
    if c.type == "tetra":
        tetra_cells = c.data
        break

if tetra_cells is None:
    raise RuntimeError("Error")

mesh_tetra = meshio.Mesh(
    points=m.points,
    cells=[("tetra", tetra_cells)],
)

meshio.write("chain.xdmf", mesh_tetra)
```
