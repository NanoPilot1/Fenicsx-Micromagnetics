```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("STT/log.txt", skiprows=1)
plt.plot(data[:,0],data[:,1], color='red', label="m_x")
plt.plot(data[:,0],data[:,2], color='blue', label="m_y")
plt.plot(data[:,0],data[:,3], color='green', label="m_z")

# Compare with nmag

data = np.loadtxt("data.txt")
plt.plot(data[0],data[1], color = 'c', linewidth=1.5, linestyle='dashed')
plt.plot(data[0],data[2], color = 'c', linewidth=1.5, linestyle='dashed')
plt.plot(data[0],data[3], color = 'c',  linewidth=1.5, linestyle='dashed')


plt.xlabel("t (ns)")
plt.ylabel("m")
plt.ylim([-0.4,0.3])
plt.legend()
plt.tight_layout()
plt.show()
```
